```python
from transformers import pipeline
from langchain_community.llms import HuggingFacePipeline

pipe = pipeline(
    "text-generation",
    model="distilgpt2",
    max_new_tokens=80
)

llm = HuggingFacePipeline(pipeline=pipe)
```

```python
response = llm.invoke("Explain LangChain in simple words.")
print(response)
```

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate.from_template(
    "Explain {topic} in simple words for a beginner."
)

formatted_prompt = prompt.format(topic="LangChain")
print(formatted_prompt)
```

```python
response2 = llm.invoke(formatted_prompt)
print(response2)
```

```python
from langchain_core.output_parsers import StrOutputParser

prompt2 = PromptTemplate.from_template(
    "Explain {topic} in 5 simple points."
)

chain = prompt2 | llm | StrOutputParser()

result = chain.invoke({"topic": "LangChain"})
print(result)
```

```python
response = llm.invoke("Write 5 short lines explaining LangChain for a beginner.")
print(response)
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate

prompt2 = PromptTemplate.from_template(
    "Give 5 short beginner-friendly points about {topic}. Keep each point on a new line."
)

chain = prompt2 | llm | StrOutputParser()

result = chain.invoke({"topic": "LangChain"})
print(result)
```

```python
chat_history = []

chat_history.append("User: My name is Rishikant.")
chat_history.append("Assistant: Nice to meet you, Rishikant.")
chat_history.append("User: What is my name?")

print("Stored Conversation History:")
for msg in chat_history:
    print(msg)
```
